2. Define a function to check whether a given number is prime or not, and display
appropriate messages.

```python
def prime(): 
    '''a function to check whether a given number is prime or not, and display appropriate messages'''
    number = int(input("Enter the number to check its prime or not: "))
    if number > 1: #prime numbers are greater than 1
     if number == 2:
      print(number,"is a prime number")
    for i in range(2,int(number/2)+1): #finding factors till number/2
     if (number%i)==0:
        print(number,"is not a prime number")
     else:
        print(number,"is a prime number")
        break
          
prime()
help(prime) #displays docstring in the output
```
