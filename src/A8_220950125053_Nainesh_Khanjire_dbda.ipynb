```python
#sf1= sf.InvestmentType.drop_duplicates(subset="Seedfunding",inplace=True)
#sf.drop(sf[(sf.InvestmentType =='SeedFunding')&(sf.InvestmentType == 'Crowd funding')&(sf.InvestmentType == 'PrivateEquity')].index.inplace=True)
sfn = sf.drop(sf[(sf['InvestmentType'] =='PrivateEquity') | (sf['InvestmentType'] =='Crowd funding')|(sf['InvestmentType'] =='SeedFunding')].index)
sfn
```

```python
sfn.info
```

```python
# visualize frequency distibution of startup industry types in india using counterplot and pie-plot
import matplotlib.font_manager as fm
#using custom fonts for titles
customfont = fm.FontProperties(fname='../data/JetBrainsMono-Medium.ttf')

f,ax = plt.subplots(1,2,figsize=(30,10))
ax[0] = sfn['InvestmentType'].value_counts().plot.pie(explode=None,shadow = True,autopct='%1.1f%%',pctdistance=0.3,labeldistance=.5,startangle =0,radius =2,ax=ax[0],legend=True)
ax[0].set_title('INVESTMENTS',fontproperties=customfont,size=10)


ax[1]= sns.countplot(x='InvestmentType',data=sfn,palette = 'Dark2')
ax[1].set_title('Investment-wise Frequency Distribution',fontproperties=customfont,size=15)

ax[0].axis('equal')
plt.show()
```

```python
 # visualize frequency distribution of `city-location` variable
    
f, ax = plt.subplots(figsize=(30,60))
ax = sns.countplot(y="CityLocation", data=sfn, palette="Dark2")
ax.set_title("Frequency distribution of Locations")
ax.set_xticklabels(sfn.CityLocation.value_counts().index, rotation=90)
plt.show()
```
